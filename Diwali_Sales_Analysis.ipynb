# Exploratory Data Analysis

# "This graph compares the total money spent by male and female customers. It helps us understand which gender contributed more to the sales during Diwali."

```python
sales_gender = df.groupby('Gender', as_index=False)['Amount'].sum()

sns.barplot(x='Gender', y='Amount', data=sales_gender, palette='pastel')
plt.title('Total Sales by Gender')
plt.tight_layout()
plt.show()
```

# "This line chart shows the average money spent by each age group. It helps us know which age group spends the most during Diwali."

```python
avg_age = df.groupby('Age Group', as_index=False)['Amount'].mean()
sns.lineplot(x='Age Group', y='Amount', data=avg_age, marker='o')
plt.title('Average Purchase Amount by Age Group')
plt.tight_layout()
plt.show()
```

# "This bar chart shows which 5 states had the highest total sales. It helps the business know where their best markets are."

```python
# Grouping and sorting top 5 states by total sales
sales_state = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(5).reset_index()

# Plotting a vertical bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='State', y='Amount', data=sales_state, palette='pastel')

plt.title('Top 5 States by Total Sales')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# "This chart shows how many orders came from each zone. It helps us find which regions are most active during Diwali sales."

```python
zone_orders = df.groupby('Zone', as_index=False)['Orders'].sum()
sns.barplot(x='Zone', y='Orders', data=zone_orders, palette='muted')
plt.title('Total Orders by Zone')
plt.tight_layout()
plt.show()
```

# "This vertical bar chart shows the number of customers from each occupation. It’s useful to see which job types contribute most to Diwali sales."

```python
sns.countplot(x='Occupation', data=df, order=df['Occupation'].value_counts().index, palette='cool')
plt.title('Customer Count by Occupation')
plt.xticks(rotation=45)  # Rotates x-axis labels to avoid overlap
plt.tight_layout()
plt.show()
```

# "This pie chart shows the percentage of total sales from each product category. It helps identify which types of products are most popular during Diwali."

```python
plt.figure(figsize=(10, 10))  # make the chart larger

df['Product_Category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colormap='Pastel1')
plt.title('Product Category Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

# "This bar chart compares total sales from married and unmarried customers. It shows which group spends more during Diwali."

```python
marital = df.groupby('Marital_Status', as_index=False)['Amount'].sum()
sns.barplot(x='Marital_Status', y='Amount', data=marital, palette='Set2')
plt.title('Total Sales by Marital Status (0=Unmarried, 1=Married)')
plt.tight_layout()
plt.show()
```

# "This histogram shows how customers are distributed by age. It helps identify which age groups are most active in shopping."
# "This histogram shows that most shoppers are between 25 and 35 years old, with fewer older customers, suggesting young adults are the main buyers during Diwali.

```python
sns.histplot(df['Age'], bins=10, kde=True, color='orange')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

# “Each dot is one customer. The more orders they place, the more they usually spend. The color shows whether they are male or female.”

```python
sns.scatterplot(x='Orders', y='Amount', data=df, hue='Gender', palette='Set1')
plt.title('Orders vs Amount')
plt.tight_layout()
plt.show()
```

# "This is a correlation heatmap. It shows how much two variables are related using numbers and colors. Red means strong relation, blue means weak or no relation. Most values here are near zero, so things like Age, Orders, and Amount don’t really affect each other."
# Range from -1 to 1
#  1.00 = Perfect positive relation (goes up together)
# -1.00 = Perfect negative relation (one goes up, other goes down)
#  0.00 = No relation at all

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
